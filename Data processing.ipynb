- Example: Temperature Data

```python
f=open("C:/Users/wnsdu/OneDrive/바탕 화면/DS_practice1_dataset/SampleTMP.txt","r")
r=f.read()
print(r)
```

```python
with open("C:/Users/wnsdu/OneDrive/바탕 화면/DS_practice1_dataset/SampleTMP.txt","r") as f:
    lines_after=f.readlines()[3:]
    data_list=[] 
    for i in lines_after:
        data_list.append(i.split())

```

```python
import pandas as pd
header=["nSeq", "DI", "PORT3_CHN1"] # 헤더이름을 리스트로 만들어줌

dataFrame=pd.DataFrame(columns=header)  #pandas에 존재하는 DataFrame이라는 데이터 정렬함수를 이용.
with open("C:/Users/wnsdu/OneDrive/바탕 화면/DS_practice1_dataset/SampleTMP.txt","r") as f:
    lines_after=f.readlines()[3:]
    for i in lines_after:
        dataFrame=dataFrame.append(pd.Series(i.split(),index=header), ignore_index=True) 
        #dataFrame의 경우 다시 해당 변수에 저장하는 과정을 넣어주어야 한다.
f.close()
# pandas의 dataFrame에서는 list형태로 자료를 관리하지 않고 Series형태로 관리하므로 pd.Series를 붙여준다.
#index에 header로 설정해줌으로써 각 index값들이 나뉘어서 header에 맞게 저장된다.
```

```python
print("Data Load by List Type")
data_list
```

```python
print("Data Load by Pandas DataFrame")
dataFrame
#좌측에 pandas자체적으로 보유하는 index가 적혀있다. 만약 상단에 ignore_index=False로 해주었다면 series생성 과정에서
#자동으로 전부 0으로 설정된 index가 출력되게 된다. 
```

## 4) XML File Processing

- Example: COVID-19 Data (COVID-19_xml_210304.xml)

```python
import xml.etree.ElementTree as ET  #xml파일 읽어오는 패키지

tree=ET.parse("C:/Users/wnsdu/OneDrive/바탕 화면/DS_practice1_dataset/xmlSample.xml")
#xml구조상 tree형태로 데이터를 저장하는 것이 좋다. ET.parse를 이용하여 불러온 데이터를 바로 parsing할 수 있다.
```

```python
root=tree.getroot()
print(type(root))
print(root.tag) #root노드가 breakfast_menu인것을 확인
print(root.attrib)
```

```python
root.getchildren() #root에 5개의 자식노드가 존재함을 확인
```

```python
print(type(root.getchildren()[0]))
print(root.getchildren()[0].tag) # root노드의 첫번째 아들노드의 tag와 그 attribute값을 얻음
print(root.getchildren()[0].attrib)
```

```python
ElementList=root.getchildren()
print(ElementList[2])
print(ElementList[2].tag)
print(ElementList[2].attrib)
```

```python
for child in root:
    print(child.tag,child.attrib)
#root노드를 순회하면서 자식노드의 tag와 attrib값을 출력
```
